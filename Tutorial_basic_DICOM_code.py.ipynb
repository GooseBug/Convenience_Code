## 기본 Code
* DICOM 파일을 보는 방법에 대한 객체
> * `dicom_Viewer`
* 객체의 주요 파라미터
> * dcm_directory_path
>> * : DICOM 파일들이 위치한 Directory
>> * 해당 디렉터리의 하위에 존재하는 모든 DICOM 파일에 대하여 파일 이름 순으로 DICOM 파일을 순서대로 가지고 온다.
> * title_color
>> * : 이미지 출력 시, 글자의 색

```python
# DICOM 파일들이 있는 디렉터리의 경로
dicom_directory_path = "./Dicom_directory"


# dicom_Viewer의 Object 생성
DV_Object = dicom_Viewer(
    dcm_directory_path = dicom_directory_path,
    title_color="limegreen"
)
```

## 1) 순차적으로 영상 출력
* Module: `print_sequentially()`
* 위 모듈을 이용하여, 3D DICOM 이미지를 순서대로 출력한다.
* 주요 파라미터
> * title_color: title의 색깔
>> Default: limegreen

```python
# dicom_Viewer를 이용하여 순차적으로 영상을 출력한다.
DV_Object.print_sequentially()
```

## 2) 원하는 Slide 순서 한장을 출력
* Module: `print_one_index_slide()`
* 위 모듈을 이용하여, 3D DICOM 이미지 중 idx에 해당하는 한 2D 슬라이드를 출력한다.
* 주요 파라미터
> * idx: 출력할 Slide의 위치
> * dicom_header_title: 출력되는 Slide에 대하여 DICOM의 주요 Header를 출력할 것인지.
>> * Default: False

```python
DV_Object.print_one_index_slide(
    idx = 40,
    dicom_header_title = True
)
```

## 3) Slide로 영상 출력
* Module: `Slide_viewer()`
* 위 모듈을 이용하여, 3D DICOM 이미지를 Slide를 이용하여 출력한다.
* 주요 파라미터
> * slide_start_point: Slide의 첫 시작 위치
>> Default: 0
> * only_slide_number: Slide의 제목으로 Slide의 번호만 출력할 것인지
>> Default: True

```python
DV_Object.Slide_viewer(
    slide_start_point = 100,
    only_slide_number = False
)
```
